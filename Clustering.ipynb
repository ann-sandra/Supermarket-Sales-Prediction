```python
dataset['Transaction date']=pd.to_datetime(dataset['Transaction date'])
```

Filtering out Transactions that happened in the last 6 months

```python
six_months_periods = [(pd.Timestamp('2022-01-01'), pd.Timestamp('2022-06-09')),
                      (pd.Timestamp('2022-06-10'), pd.Timestamp('2022-12-09'))]
```

```python
for period_start, period_end in six_months_periods:
    period_dataset = dataset[
        (dataset['Transaction date'] >= period_start) & (dataset['Transaction date'] <= period_end)
    ]
```

```python
period_dataset
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
selected_features = ['Product category','Product subcategory','Product score 1','Product score 2','Total amount per unit','Product rating']
module3 = period_dataset[selected_features]

module3
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
categorical_columns = ['Product category', 'Product subcategory']
numerical_columns = ['Product score 1', 'Product score 2', 'Total amount per unit', 'Product rating']

# One-hot encode categorical columns
encoder = OneHotEncoder(drop='first', sparse=False)
categorical_encoded = encoder.fit_transform(period_dataset[categorical_columns])

# Combine one-hot encoded categorical columns with numerical columns
module3 = pd.concat([pd.DataFrame(categorical_encoded, columns=encoder.get_feature_names_out(categorical_columns)), period_dataset[numerical_columns]], axis=1)
```

```python
module3 = module3.dropna()
scaler = StandardScaler()
module3_scaled = scaler.fit_transform(module3)

```

```python
inertia = []
for k in range(1, 11):  
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(module3_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

The elbow curve shows the inertia (within-cluster sum of squares) of the k-means algorithm for different values of k. The inertia is a measure of how well the data is grouped into clusters. A lower inertia indicates a better clustering.

The elbow curve shows that the inertia decreases sharply as k increases from 2 to 5. After k=5, the inertia decreases more slowly. This suggests that k=5 is the optimal number of clusters for this data.

As the inertia decreases by 200 points from k=4 to k=5, but only by 30 points from k=5 to k=6. Therefore, the elbow is taken to be at k=5.

```python
categorical_columns = ['Product category', 'Product subcategory']
numerical_columns = ['Product score 1', 'Product score 2', 'Total amount per unit', 'Product rating']

# One-hot encode categorical columns
encoder = OneHotEncoder(drop='first', sparse=False)
categorical_encoded = encoder.fit_transform(period_dataset[categorical_columns])

# Combine one-hot encoded categorical columns with numerical columns
module3 = pd.concat([pd.DataFrame(categorical_encoded, columns=encoder.get_feature_names_out(categorical_columns)), period_dataset[numerical_columns]], axis=1)

# Drop rows with missing values
module3 = module3.dropna()

# Standardize the data
scaler = StandardScaler()
module3_scaled = scaler.fit_transform(module3)

# Choose the number of clusters (you can modify this based on the elbow method)
num_clusters = 5

# Apply KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
module3['Cluster'] = kmeans.fit_predict(module3_scaled)

# Extract one-hot encoded column names
one_hot_encoded_columns = encoder.get_feature_names_out(categorical_columns).tolist()

# Combine one-hot encoded columns with numerical columns and 'Cluster' column
selected_features_with_cluster = numerical_columns + one_hot_encoded_columns + ['Cluster']

# Group by 'Cluster' and calculate the mean of selected features
cluster_summary = module3.groupby('Cluster')[selected_features_with_cluster].mean()

# Plot pairplot
sns.pairplot(module3, hue='Cluster', vars=numerical_columns)
plt.show()

# Display cluster summary
print("Cluster Summary:")
print(cluster_summary)
```

*INTERPRETATION*

**Cluster 0:**

High-End Buyers with Moderate Product Rating: This cluster represents products with high product score 1 (51.31), slightly below average product score 2 (-1.20), and high total amount per unit (30.29). They might be willing to pay more for a product that meets their basic needs but may not be overly concerned with aesthetics or additional features.
